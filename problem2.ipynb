```python
import numpy as np
import matplotlib.pyplot as plt


A = np.array([[3.0, 0.5],[0.5, 1.0]])
mu = np.array([1, 2])

rate = 1/np.max(np.linalg.eig(2*A)[0])

lambs = np.array([2, 4, 6])
w = np.array([3, -1],dtype=float)
arr=[]
for lamb in lambs:
    for i in range(0,100):
        temp0 = w[0]-rate*(2*np.dot((w-mu), A)[0])
        temp1 = w[1]-rate*(2*np.dot((w-mu), A)[1])
        a=rate*lamb
        if temp0>a:
            w[0] =temp0-a
        elif temp0<-a:
            w[0] =temp0+a
        else:
            w[0]= 0

        if temp1>a:
            w[1] =temp1-a
        elif temp1<-a:
            w[1] =temp1+a
        else:
            w[1]= 0
    arr.append(w[0])
    arr.append(w[1])
    
iteration = []
for i in range(100):
    iteration.append(i)
cnt=0
#λ=2
w = np.array([3, -1],dtype=float)
diff_2 = []

for i in range(0,100):
    temp0 = w[0]-rate*(2*np.dot((w-mu), A)[0])
    temp1 = w[1]-rate*(2*np.dot((w-mu), A)[1])
    a=rate*2
    if temp0>a:
        w[0] =temp0-a
    elif temp0<-a:
        w[0] =temp0+a
    else:
        w[0]= 0

    if temp1>a:
        w[1] =temp1-a
    elif temp1<-a:
        w[1] =temp1+a
    else:
        w[1]= 0
    diff_2.append(np.linalg.norm(w-[round(arr[0],2), round(arr[1],2)], ord = 2))
    
#λ=4
w = np.array([3, -1],dtype=float)
diff_4 = []
for i in range(0,100):
    temp0 = w[0]-rate*(2*np.dot((w-mu), A)[0])
    temp1 = w[1]-rate*(2*np.dot((w-mu), A)[1])
    a=rate*4
    if temp0>a:
        w[0] =temp0-a
    elif temp0<-a:
        w[0] =temp0+a
    else:
        w[0]= 0

    if temp1>a:
        w[1] =temp1-a
    elif temp1<-a:
        w[1] =temp1+a
    else:
        w[1]= 0
    diff_4.append(np.linalg.norm(w-[round(arr[2],2), round(arr[3],2)], ord = 2))
    
    
    
#λ=6
w = np.array([3, -1],dtype=float)
diff_6 = []
for i in range(0,100):
    temp0 = w[0]-rate*(2*np.dot((w-mu), A)[0])
    temp1 = w[1]-rate*(2*np.dot((w-mu), A)[1])
    a=rate*6
    if temp0>a:
        w[0] =temp0-a
    elif temp0<-a:
        w[0] =temp0+a
    else:
        w[0]= 0

    if temp1>a:
        w[1] =temp1-a
    elif temp1<-a:
        w[1] =temp1+a
    else:
        w[1]= 0
    diff_6.append(np.linalg.norm(w-[round(arr[4],2), round(arr[5],2)], ord = 2))
    
plt.figure()
plt.yscale('log')
plt.plot(iteration, diff_2,label='λ=2')
plt.plot(iteration, diff_4,label='λ=4')
plt.plot(iteration, diff_6,label='λ=6')
plt.xlabel('iteration')
plt.ylabel('||w(t) - w^||')
plt.legend()


```
